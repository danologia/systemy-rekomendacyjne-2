## Frequent Itemsets via Apriori Algorithm

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
%time
frequent_itemsets = apriori(df, min_support=0.0004, low_memory=True, use_colnames=True)
len(frequent_itemsets)
# min_support=0.0004  => 5103
# min_support=0.00036  => ?
# min_support=0.0003  => 621567
```

```python
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets = frequent_itemsets[frequent_itemsets['length'] >= 2].reset_index(drop=True)
frequent_itemsets.sort_values(by=['support'], ascending=False)
```

```python
# biggest support 0.001776 frequent_itemsets pair (174086, 126335)
df_train_items = pd.read_csv('data/dataset_split/train.csv')
df_train_items[df_train_items['item_id'] == 174086].head()
```

```python
df_train_items[df_train_items['item_id'] == 126335].head()
```

```python
itemsets = frequent_itemsets['itemsets'].tolist()
support_list = frequent_itemsets['support'].tolist()
itemsets[:5]
```

```python
def predict_items(item):
    pred = {}
    for itemset, support in zip(itemsets, support_list):
        if item in itemset:
            new_items = list(itemset - {item})
            for new_item in new_items:
                pred[new_item] = support
#             pred = pred + list(itemset - {item})
    return pred

predict_items(123793)
```

```python
def get_user_items(user_id):
    return df_items[df_items['user_id'] == user_id]['items'].to_list()[0]

def get_user_recommendation(user_id):
    predicted_items = [predict_items(item_id) for item_id in get_user_items(user_id)]
    items_dict = {}
    for p in predicted_items:
        items_dict = {**items_dict, **p}
    return items_dict

get_user_recommendation(332)
```
